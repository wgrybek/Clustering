```python
from sklearn_extra.cluster import KMedoids
from sklearn.datasets import make_blobs, make_moons, make_circles
import matplotlib.pyplot as plt
```

# Blobs

```python
X, y_true = make_blobs(n_samples=600, centers=3, cluster_std=2, random_state=1234)
```

```python
kmedoid = KMedoids(n_clusters=3)
kmedoid.fit(X)
y_kmedoid = kmedoid.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmedoid, s=20, cmap='Set3_r')
centers = kmedoid.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=50, alpha=0.9);
plt.show()
```

# Księżyce

```python
X, y = make_moons(400, noise=.05, random_state=0)
kmeans = KMedoids(n_clusters=2)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=20, cmap='Set3_r')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=20, alpha=0.9)
plt.show()
```

# Koła

```python
X, y = make_circles(600,factor = 0.5, noise=0.05, random_state=0)
kmeans = KMedoids(n_clusters=2)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=20, cmap='Set3_r')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=20, alpha=0.9)
plt.show()
```
